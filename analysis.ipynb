```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("MOE kenyan schools.csv")
df
```

```python
row_drop_2=["NGO/CBO(Plan Interntnl)","MIN.GENDER, SPORTS ,CULTURE","RCO","OTHER,MINISTRY OF STATE YOUTH AFFAIRS","PTA","FATHER GROLL'S WELFAE PROJECT","PUBLIC","KENYA NAVY","OTHER, MINISTRY OF YOUTH AFFAIRS","LL","ESTATE-KIRIMARA FARM","ESTATE","DOD","DEB","NON -PROFIT MISSION UNDER SOCIETY ACT","NON-PROFIT MISSION UNDER SOCIETY ACT","PRIVATE INDIVIDUAL/PRIVATE ORGANIZATION"]
mask=df["Sponsor"].isin(row_drop_2)

df=df[~mask]
df
```

```python
df["Sponsor"].value_counts()
```

```python
row_drop_3=["OTHER, POLICE","OTHER, MINISTRY OF STATE YOUTH AFFAIRS","MIN.GENDER, SPORTS, CULTURE","KAGANDA PARENTS","CI","COP","GK PRISON ELDORERT","E"]
mask=df["Sponsor"].isin(row_drop_3)

df=df[~mask]
df
```

```python
df["Sponsor"].value_counts()
```

```python
row_drop_4=["GK PRISON ELDORET","OTHER, POLICE","NONE"]
mask=df["Sponsor"].isin(row_drop_4)

df=df[~mask]
df
```

```python
df["Sponsor"].value_counts()
```

```python
df.info()
```

```python
df.drop(["COSTITUENCY", "DIVISION", "LOCATION", "SUBLOCATIO","# Classrooms"], axis=1, inplace=True)
```

```python
df
```

```python
df.info()
```

```python
mode_level=df["Level"].mode().iloc[0]
mode_level
```

```python
df["Level"].fillna(mode_level,inplace=True)
```

```python
mode_status=df["Status"].mode().iloc[0]
mode_status
```

```python
df["Status"].fillna(mode_status,inplace=True)
```

```python
mode_sponsor=df["Sponsor"].mode().iloc[0]
mode_sponsor
```

```python
df["Sponsor"].fillna(mode_sponsor,inplace=True)
```

```python
df.info()
```

```python
df["Province_1"]=df["PROVINCE"].astype("category").cat.codes
df["District_1"]=df["DISTRICT"].astype("category").cat.codes
df["level_1"]=df["Level"].astype("category").cat.codes
df["status_1"]=df["Status"].astype("category").cat.codes
df["sponsor_1"]=df["Sponsor"].astype("category").cat.codes
```
